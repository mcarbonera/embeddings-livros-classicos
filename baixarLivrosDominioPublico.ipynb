# Script para download de livros de domínio público

Este código realiza o download de livros disponibilizados no site www.netmundi.org

## Imports

```python
#%pip install bs4
#%pip install requests
```

```python
from bs4 import BeautifulSoup as bs
import requests
import os
import re
```

## Acessar a página com links para livros

```python
header = {
  'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36',
}
URL = 'https://www.netmundi.org/home/classicos-da-literatura-e-da-filosofia-300-obras-para-baixar/'

r = requests.get(URL, headers=header)
soup = bs(r.text)
```

## Encontrar URLs

```python
texto = soup.find_all('li')
urls = []
names = []

for tag in texto:
  for i, link in enumerate(tag.find_all('a')):
    FULLURL = link.get('href')
    if(str(FULLURL).find('/uploads') > -1):
      urls.append(FULLURL)
      name = tag.select('a')[i].contents[0] + FULLURL.split('/')[-1]
      name = re.sub("[:?]", "", name)
      names.append(name)
```

## Download dos arquivos

```python
novaPasta = 'livros'
if not os.path.exists('livros'):
  os.makedirs(novaPasta)
```

```python
names_urls = zip(names, urls)
tamanho = len(names)
num = 1

for name, url in names_urls:
  response = requests.get(url, headers=header)
  pdf = open("livros/" + name, 'wb')
  pdf.write(response.content)
  pdf.close()
  print('Download '+str(num)+'/'+str(tamanho)+': OK')
  num=num+1
```
